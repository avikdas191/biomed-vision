```python
from dicom_parser import Image
```

```python
imgD = Image("/kaggle/input/datasets/avikdas191/biomedical-computer-vision/image-00178.dcm")
ds = imgD.header.raw
```

```python
ds
```

```python
# What imaging technique was used on the patient?
print(ds.get((0x0008, 0x0060)).value) # Modality

# On which body part?
print(ds.get((0x0008, 0x1030)).value) # Study Description
```

```python
# When was the image taken?
print(ds.get((0x0008, 0x0023)).value) # Content Date
```

```python
# What is the patient position in the CT machine?
patient_position = ds.get((0x0018, 0x5100)).value
print(patient_position) # Patient Position

# What would the code for the opposite orientation be?
opposites = { # Define opposite mapping
    "FFS": "HFP",
    "HFP": "FFS",
    "HFS": "FFP",
    "FFP": "HFS"
}
opposite_position = opposites.get(patient_position)
print(opposite_position)
```

```python
# What is the xyz image size in millimeters?
rows = ds.get((0x0028, 0x0010)).value
cols = ds.get((0x0028, 0x0011)).value
pixel_spacing = ds.get((0x0028, 0x0030)).value
slice_thickness = ds.get((0x0018, 0x0050)).value

x_mm = cols * pixel_spacing[1]
y_mm = rows * pixel_spacing[0]
z_mm = slice_thickness

print(f"X: {x_mm:.2f} mm")
print(f"Y: {y_mm:.2f} mm")
print(f"Z: {z_mm:.2f} mm")
```

```python
# What image compression format is used for the pixel data? 
print(ds.file_meta.TransferSyntaxUID.name)

# What is the bit depth?
print(ds.get((0x0028, 0x0101)).value)
```

```python
# What is the tag for the patient name? 
print(ds.get((0x0010, 0x0010)).value)

# Do we have that information? NO
```
