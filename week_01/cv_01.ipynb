```python
from dicom_parser import Image
```

```python
imgD = Image("/kaggle/input/datasets/avikdas191/biomedical-computer-vision/image-00178.dcm")
```

```python
imgD.header.raw
```

```python
# What imaging technique was used on the patient? 
print(imgD.header.raw["Modality"].value)

# On which body part?
print(imgD.header.raw["StudyDescription"].value)
```

```python
# When was the image taken?
print(imgD.header.raw["ContentDate"].value)
```

```python
# What is the patient position in the CT machine?
print(imgD.header.raw["PatientPosition"].value)

# What would the code for the opposite orientation be?
```

```python
# What is the xyz image size in millimeters?
# print(img.header.raw["Rows"].value)
# print(img.header.raw["Columns"].value)
rows = imgD.header.raw["Rows"].value
cols = imgD.header.raw["Columns"].value
spacing = imgD.header.raw["PixelSpacing"].value
thickness = imgD.header.raw["SliceThickness"].value

size_x_mm = cols * spacing[1]
size_y_mm = rows * spacing[0]
size_z_mm = thickness

print(size_x_mm)
print(size_y_mm)
print(size_z_mm)
```
